```python
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import nltk
```

```python
sentence = [['king', 'queen', 'man', 'women'],
            ['paris', 'france', 'rome', 'italy'],
            ['apple', 'orange', 'banana', 'grapes']]

model = Word2Vec(sentence,
                 vector_size=100,
                 window=5,
                 min_count=1,
                 workers=4)
vector = model.wv['king']
print(vector)
```

```python
sentences = [
    ["king", "queen", "man", "woman", "prince", "princess", "royal", "palace"],
    ["doctor", "nurse", "hospital", "medicine"],
    ["paris", "france", "berlin", "germany", "rome", "italy"],
    ["apple", "banana", "orange", "fruit"],
    ["car", "truck", "vehicle", "road"]]

model = Word2Vec(sentences,
                 vector_size=20,
                 window=2,
                 min_count=1,
                 workers=4, seed=42)
```

```python
words =[ "king", "queen", "man", "woman", "prince", "princess", "palace",
         "paris", "france", "berlin", "germany", "rome", "italy",
         "doctor", "nurse", "hospital", "medicine",
         "apple", "banana", "fruit", "car", "truck", "vehicle"]

word_vectors = [model.wv[word] for word in words]
```

```python
pca = PCA(n_components=2)
reduced_vectors = pca.fit_transform(word_vectors)
```

```python
plt.figure(figsize=(10, 7))
for i, word in enumerate(words):
    plt.scatter(*reduced_vectors[i], label=word)
    plt.text(reduced_vectors[i][0]+0.005, reduced_vectors[i][1]+0.005, word, fontsize=10)

plt.title('Word2Vec - 2D visualisation')
plt.grid(True)
plt.tight_layout()
plt.show()
```
